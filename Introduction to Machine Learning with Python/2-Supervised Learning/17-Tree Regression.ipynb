```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ram = pd.read_csv('ram_price.csv')
```

```python
plt.figure(figsize=(7, 5))
plt.semilogy(ram.date, ram.price)
plt.xlabel("Year")
plt.xlabel("Price in $/Mbytes")
```

```python
train = ram[ram.date < 2000]
test = ram[ram.date >= 2000]
```

```python
X_train = train.date[:, np.newaxis]
y_train = np.log(train.price)
```

```python
tree = DecisionTreeRegressor().fit(X_train, y_train)
linear = LinearRegression().fit(X_train, y_train)
```

```python
X = ram.date[:, np.newaxis]
```

```python
yhat_tree = tree.predict(X)
yhat_linear = linear.predict(X)
```

```python
plt.figure(figsize=(10, 10))
plt.semilogy(train.date, train.price, label="Training Data")
plt.semilogy(test.date, test.price, label="Test Data")
plt.semilogy(ram.date, yhat_tree, label="Tree Prediction")
plt.semilogy(ram.date, yhat_linear, label="Linear Prediction")
plt.legend()
```
