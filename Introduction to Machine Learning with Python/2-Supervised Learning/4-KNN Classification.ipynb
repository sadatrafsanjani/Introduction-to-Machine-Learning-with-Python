```python
import mglearn
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

```python
X,y = make_blobs(n_samples=100, n_features=2, random_state=0)
plt.figure(figsize=(7,7))
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['Class 0', 'Class 1'], loc=3)
plt.xlabel('First Feature')
plt.ylabel('Second Feature')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n,ax in zip([1,3,9], axes):
    clf = KNeighborsClassifier(n_neighbors=n).fit(X, y)
    mglearn.plots.plot_2d_seperator(clf, X, fill=True, eps=0.5, ax=ax, alpha=0.8)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n))
    plt.set_xlabel('First Feature')
    plt.set_ylabel('Second Feature')
axes[0].legend(loc=3)
```
