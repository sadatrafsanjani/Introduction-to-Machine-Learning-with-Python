```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import mglearn
import matplotlib.pyplot as plt
```

```python
X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify=y, random_state=42)
```

```python
def plotter(mlp):

    mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=0.3)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
    plt.xlabel("Feature 0")
    plt.xlabel("Feature 1")
```

```python
mlp = MLPClassifier(activation='relu',random_state=0, max_iter=2000)
mlp.fit(X_train, y_train)
```

```python
plotter(mlp)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,10), activation='relu',random_state=0, max_iter=2000)
mlp.fit(X_train, y_train)
```

```python
plotter(mlp)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,10), activation='tanh',random_state=0, max_iter=2000)
mlp.fit(X_train, y_train)
```

```python
plotter(mlp)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,10,100,50), activation='relu',random_state=0, max_iter=2000)
mlp.fit(X_train, y_train)
```

```python
plotter(mlp)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 8))

for axx, n_nodes in zip(axes, [10, 100]):
    
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        
        mlp = MLPClassifier(hidden_layer_sizes=(n_nodes, n_nodes), random_state=0, alpha=alpha, max_iter=2000)
        mlp.fit(X_train, y_train)
        
        mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)
        mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
        ax.set_title("n_hidden=[{}, {}]\nalpha={:.4f}".format(n_nodes, n_nodes, alpha))
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 8))

for i, ax in enumerate(axes.ravel()):
    
    mlp = MLPClassifier(hidden_layer_sizes=(100, 100), random_state=i, max_iter=2000)
    mlp.fit(X_train, y_train)
    
    mglearn.plots.plot_2d_separator(mlp, X_train, fill=True, alpha=.3, ax=ax)
    mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train, ax=ax)
```
