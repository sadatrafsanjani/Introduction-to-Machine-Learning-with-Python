```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import mglearn
```

```python
dataset = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset['data'], dataset['target'], random_state=0)
```

```python
print("X_train Shape: {}".format(X_train.shape))
print("y_train Shape: {}".format(y_train.shape))
```

```python
print("X_test Shape: {}".format(X_test.shape))
print("y_test Shape: {}".format(y_test.shape))
```

```python
df = pd.DataFrame(X_train, columns=dataset.feature_names)
grr = pd.plotting.scatter_matrix(df, c=y_train, alpha=0.8, figsize=(15, 15), marker='o', hist_kwds={'bins':20})
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
X_new = np.array([[1.2, 5.4, 2.5,7.2]])
print("X_train Shape: {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("Prediction: {}".format(dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("Test Set Prediction: {}".format(dataset['target_names'][y_pred]))
```

```python
print("Test Set Score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Test Set Score: {:.2f}".format(knn.score(X_test, y_test)))
```
